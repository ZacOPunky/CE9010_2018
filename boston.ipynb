```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
# Read the table and print out the data
data = pd.read_csv('data/train.csv')
print(data.shape)
print(data.ndim)
# Find the mean and variance of each column
print(np.mean(data,axis=0))
print(np.var(data,axis=0))
```

```python
#Normalise the data 
bosdf=data
bosdf -= bosdf.mean(axis=0)
bosdf /= np.std(bosdf)
print(np.mean(bosdf))
print(np.std(bosdf))
print(np.var(bosdf))
print(bosdf)
```

```python
temp=bosdf.iloc[:].values
bosdfnorm=np.zeros((data.shape[0],data.shape[1]+1),dtype=float)
for i in range(1,data.shape[0]+1):
    bosdfnorm[i-1,:]=i

print(temp.shape)
print(bosdfnorm.shape)
bosdfnorm[:,1:]=temp[:,:]
fig=plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(111)
ax1.scatter(bosdfnorm[:,6], bosdfnorm[:,11],color='red', label='rm vs ptratio');


plt.legend()
plt.xlim(-2.5,3)
plt.ylim(-2.5,2.5)
```
